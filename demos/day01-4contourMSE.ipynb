```python
X = np.ones((10,2))
X[:,0] = x
print(X)
```

```python
A = np.matmul(X.T,X)
A
```

```python
b = np.matmul(X.T, y)
print(b)
```

```python
wopt = np.linalg.solve(A, b)
print(wopt)
```

```python
yh = np.matmul(X, wopt)
dy = yh - y
dy2 = dy*dy
rmse = np.sqrt(dy2.mean())
print('RMSE = ', rmse)
```

This is how well our linear predictor is doing.
