## Lession 8 - Data Normalization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
%matplotlib inline
```

```python
# inputs and corresponding outputs (targets)
x = np.array([67,67,72,69,69,70,67,67,66,68])
y = np.array([128,123,129,143,132,142,112,118,108,119])
```

```python
#definition of MSE loss function
def MSE(w, b, x, y):
    yh = w*x+b
    dy = y - yh
    dy2 = dy*dy
    return dy2.mean()
```

```python
#define contour plot
def plot_RMSE(wmin, wmax, bmin, bmax, num_pts, x, y):
    rmse = np.zeros((num_pts, num_pts))
    ww, bb = np.meshgrid(np.linspace(wmin, wmax, num_pts), np.linspace(bmin, bmax, num_pts))
    for i in range(num_pts):
        for j in range(num_pts):
            rmse[i, j] = np.sqrt(MSE(ww[i, j], bb[i, j], x, y))
    plt.contourf(ww, bb, rmse, 25)
    plt.show()
```

```python
plot_RMSE(0, 10, -180, -160, 100, x, y)
```

```python
x_s = (x-x.mean())/x.std()
y_s = (y-y.mean())/y.std()
```

```python
plot_RMSE(-1, 1, -1, 1, 100, x_s, y_s)
```
