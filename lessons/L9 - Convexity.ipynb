### Plot the contours of the function.

```python
#define contour plot
def plot_fa(x1min, x1max, x2min, x2max, num_pts):
    fa = np.zeros((num_pts, num_pts))
    x1, x2 = np.meshgrid(np.linspace(x1min, x1max, num_pts), np.linspace(x2min, x2max, num_pts))
    for i in range(num_pts):
        for j in range(num_pts):
            fa[i, j] = x1[i, j] + x1[i, j]*x2[i, j] + x2[i, j]**2
    plt.contour(x1, x2, fa, 25)
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.show()
```

```python
plot_fa(-20, 20, -20, 20, 100)
```

```python
#define contour plot
def plot_fb(x1min, x1max, x2min, x2max, num_pts):
    fb = np.zeros((num_pts, num_pts))
    x1, x2 = np.meshgrid(np.linspace(x1min, x1max, num_pts), np.linspace(x2min, x2max, num_pts))
    for i in range(num_pts):
        for j in range(num_pts):
            fb[i, j] = x1[i, j]**2 + x1[i, j]*x2[i, j] + x2[i, j]**2
    plt.contour(x1, x2, fb, 25)
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.show()
```

```python
plot_fb(-20, 20, -20, 20, 100)
```

### Plot the value of the function along a line segment in the domain of the function

p1 is the starting point and p2 the ending point

```python
num_points = 10
p1 = np.array([-20,  10])
p2 = np.array([ 20, -10])
a = np.linspace(0, 1, num_points)  # alpha
fa = np.zeros(num_points)
fb = np.zeros(num_points)
for k in range(0, num_points):
    x1, x2 = a[k]*p1 + (1-a[k])*p2
    fa[k] = x1    + x1*x2 + x2**2 
    fb[k] = x1**2 + x1*x2 + x2**2 
plt.plot(a, fa, label=r'$x_1*x_1x_2+x_2^2$')
plt.plot(a, fb, label=r'$x_1^2*x_1x_2+x_2^2$')
plt.xlabel(r'$\alpha$')
plt.ylabel('fa and fb')

plt.legend()
# plt.grid(True)
plt.show()
```
