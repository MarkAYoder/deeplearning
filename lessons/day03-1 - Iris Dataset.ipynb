Example 39 - Iris Classification

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/Iris-cleaned.csv')
df.head(3)
```

```python
df = pd.get_dummies(df)
df.head(3)
```

index with iloc.   Need to add bias

```python
X = df[['sepal length', 'sepal width']]
X['bias'] = 1
X = X.values
Y = df.iloc[:,4:].values
Y
```

```python
A = np.matmul(X.T, X)
B = np.matmul(X.T, Y)
Wopt = np.linalg.solve(A, B)
Wopt
```

Rows are for features and cols are for targets

```python
Yh = np.matmul(X, Wopt)
np.round(Yh, 1).argmax(axis=1)
```
